#### Unpivot the data and clean

We will unpivot the individual dataframes to bring them in long formats and clean them before merging. <br/><br/>
We'll perform following tasks
<ul>
    <li>Since the data is in wide format i.e. individual columns for dates so we'll first bring it in long format i.e unpivot it using a function melt() in pandas</li>
    <li>The dates are not in correct formats i.e. they are in string formats so we'll convert them in date format</li>
    <li>Remove data of cruise ships "Grand Princess", "Diamond Princess" and "MS Zaandam" as that would need different kind of cleaning considering the 'Province/State' and 'Country/Region' mismatch over cruise sailing time</li>
    <li>Look for missing values and treat wherever possible and applicable</li>
    <li>Aggregate the data at Country/Region and Date level. Since Province/State data is not available for most of the countries so we'll drop that column</li>
</ul>

```python
dates = confirmed_df.columns[4:]
```

#### Confirmed Cases DataFrame 

```python
# Unpivot the dataframe, format date and remove cruise ship data
confirmed_df_long = confirmed_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=dates,
    var_name='Date',
    value_name='Confirmed'
)

confirmed_df_long['Date'] = pd.to_datetime(confirmed_df_long['Date'])
confirmed_df_long.drop(confirmed_df_long[confirmed_df_long['Country/Region'].str.contains('Grand Princess') | confirmed_df_long['Country/Region'].str.contains('Diamond Princess') | confirmed_df_long['Country/Region'].str.contains('MS Zaandam') | confirmed_df_long['Province/State'].str.contains('Grand Princess') | confirmed_df_long['Province/State'].str.contains('Diamond Princess') | confirmed_df_long['Province/State'].str.contains('MS Zaandam')].index,inplace=True)
```

```python
# Check for NaN or missing data.

confirmed_df_long.isna().sum()
```

<b>Cofirmed</b> column and <b>Country/Region</b> column don't have missing values which is great. <b>Province/State</b> and <b>Lat/Long</b> columns can have missing values as many countries have not reported them properly. Anyway we will drop these columns post aggregation of our data.

Now Let's group our data at <b>Country/Region</b> and <b>Date</b> level.

```python
confirmed_df_long_grouped = confirmed_df_long.groupby(['Date','Country/Region'])['Confirmed'].sum().reset_index()
```

```python
confirmed_df_long_grouped
```

#### Deaths Cases DataFrame 

```python
deaths_df_long = deaths_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=dates,
    var_name='Date',
    value_name='Deaths'
)

deaths_df_long['Date'] = pd.to_datetime(deaths_df_long['Date'])
deaths_df_long.drop(deaths_df_long[deaths_df_long['Country/Region'].str.contains('Grand Princess') | deaths_df_long['Country/Region'].str.contains('Diamond Princess') | deaths_df_long['Country/Region'].str.contains('MS Zaandam') | deaths_df_long['Province/State'].str.contains('Grand Princess') | deaths_df_long['Province/State'].str.contains('Diamond Princess') | deaths_df_long['Province/State'].str.contains('MS Zaandam')].index,inplace=True)
```

```python
# Check for NaN or missing values

deaths_df_long.isna().sum()
```

There are no missing values in Deaths column so no need of cleanup.

Aggregate the data at <b>Country/Region</b> and <b>Date</b> level.

```python
deaths_df_long_grouped = deaths_df_long.groupby(['Date','Country/Region'])['Deaths'].sum().reset_index()
```

```python
deaths_df_long_grouped
```

#### Recovered Cases DataFrame 

```python
recovered_df_long = recovered_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=dates,
    var_name='Date',
    value_name='Recovered'
)

recovered_df_long['Date'] = pd.to_datetime(recovered_df_long['Date'])
recovered_df_long.drop(recovered_df_long[recovered_df_long['Country/Region'].str.contains('Grand Princess') | recovered_df_long['Country/Region'].str.contains('Diamond Princess') | recovered_df_long['Country/Region'].str.contains('MS Zaandam') | recovered_df_long['Province/State'].str.contains('Grand Princess') | recovered_df_long['Province/State'].str.contains('Diamond Princess') | recovered_df_long['Province/State'].str.contains('MS Zaandam')].index,inplace=True)
```

```python
recovered_df_long.isna().sum()
```

```python
recovered_df_long_grouped = recovered_df_long.groupby(['Date','Country/Region'])['Recovered'].sum().reset_index()
```

```python
recovered_df_long_grouped
```

#### Merge 3 dataframes into one 

```python
# Merging confirmed and deaths dataframes....and then resulting dataframe with recovered dataframe.

full_table = confirmed_df_long_grouped.merge(right=deaths_df_long_grouped,how='left',on=['Country/Region', 'Date'])
full_table = full_table.merge(right=recovered_df_long_grouped,how='left',on=['Country/Region','Date'])
full_table
```

```python
full_table.isna().sum()
```

### Add columns for Daily new cases 

We can add <b>New confirmed cases</b>, <b>New deaths</b> and <b>New recovered</b> columns as well in this dataframe. All we need to do is deduct corresponding cumulative data on the previous day from the day in consideration. 

```python
# New Confirmed Cases, New Deaths, New Recovered
temp = full_table.groupby(['Country/Region','Date'])['Confirmed','Deaths','Recovered']
temp = temp.sum().diff().reset_index()
temp
```

The above piece of code has calculated difference from date in consideration to previous day...but there will be problem where Country/Region Changes so we'll have to write code to treat that.

```python
# Masking all rows where a new country starts otherwise it will have undesired values.
mask = temp['Country/Region'] != temp['Country/Region'].shift(1)

temp.loc[mask, 'Confirmed'] = np.nan
temp.loc[mask, 'Deaths'] = np.nan
temp.loc[mask, 'Recovered'] = np.nan
```

```python
# Renaming columns to match with our main dataframe 
temp.columns = ['Country/Region','Date','New Cases','New Deaths','New Recovered']

# Merging with main dataframe....if condition is to stop it from adding more columns if the cell is run more than once
if 'New Cases' not in full_table.columns:
    full_table = pd.merge(full_table,temp,how='inner',on=['Country/Region','Date'])
full_table
```

```python
# Replace NaN values with 0 in every columns
full_table = full_table.fillna(0)

# Fix DataTypes of newly added columns as int
cols = ['New Cases', 'New Deaths', 'New Recovered']
full_table[cols] = full_table[cols].astype(int)

# To take care of negative values in New Cases column if any
full_table['New Cases'] = full_table['New Cases'].apply(lambda x : 0 if x<0 else x)
```

```python
full_table
```

```python
# Finally one more..not necessary though....Updating country name 'Korea, South' to 'South Korea' 

full_table['Country/Region'] = full_table['Country/Region'].replace(['Korea, South'],'South Korea')
```

#### Output to CSV 

```python
full_table.to_csv("covid_19_data.csv",index=False)
```
